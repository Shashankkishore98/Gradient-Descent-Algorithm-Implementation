```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function body for implementing Gradient Descent Algorithm
def gda(w0,w1,alpha,df):
    
    mse_=[] # for storing mean_suared_error values for every iteration
    w1_=[] # for storing slope values for every iteration
    w0_=[] # for storing intercept values for every iteration
    
    for i in range(50): # doing 50 iterations
        
        df['Y_pred'] = df['X1']*w1 + w0  # calculating the y_pred values
        
        e1 = np.sum(df['Y_pred']-df['Y']) # differential error wrt intercept
        
        e2=  np.sum((df['Y_pred']-df['Y'])*df['X1']) # differential error wrt slope
        
        mse = 1/df.shape[0]*np.sum((df['Y_pred']-df['Y'])**2)  # calculating the mse value
        
        
        
        # appending the respective metrics
        mse_.append(mse)
        w1_.append(w1)
        w0_.append(w0)
        
        
        # updating the slope and intercept
        w1 = w1- alpha*e2
        w0 = w0 - alpha*e1
        
        # plotting to see how the predicted line becomes best fit line and the error is minimising
        print("slope: {} ,intercept :{} intercept ,mse: {}".format(w1,w0,mse))
        plt.scatter(df['X1'], df['Y']) 
        plt.plot([min(df['X1']), max(df['X1'])], [min(df['Y_pred']), max(df['Y_pred'])], color='red')
        plt.show()
        
    return w1_,w0_,mse_
        
        
        
    
```

```python
w0,w1 =0,0
alpha =0.01
df = {'X1':[1,3,5],'Y':[4.8,12.4,15.5]}
df = pd.DataFrame(df)
a=gda(w0,w1,alpha,df)
```

```python
plt.plot(a[0],a[2])
plt.title("Loss function  vs w1 (slope)")
plt.xlabel("slope")
plt.ylabel("Cost_function")
plt.show()
```

```python
plt.plot(a[1],a[2])
plt.title("Loss function  vs w0 (intercept)")
plt.xlabel("slope")
plt.ylabel("Cost_function")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection ='3d')
ax.set_title("3d plot for cost_function wrt slope and intercept")
ax.plot(a[0],a[1],a[2])
ax.set_xlabel('slope')
ax.set_ylabel('intercept')
ax.set_zlabel('cost_function')
```
